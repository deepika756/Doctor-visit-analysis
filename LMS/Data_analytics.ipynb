# Visualize the percentage of people getting govt health insurance due to low income

```python
label=['yes', 'no']

Y=df[df['freepoor']=='yes'] 
N=df[df['freepoor']=='no' ]

x=[Y.shape[0],N.shape[0]]

plt.figure(figsize=(5,5))

plt.pie(x, labels=label)

plt.title("% of people getting govt health insurance due to low income ")

plt.show()
```

# Plot horizontal barchart to analyse the reduced days of activity due to illness based on Gender

```python
db=df.groupby('gender')['reduced'].sum().to_frame().reset_index() 
    
plt.barh(db['gender'], db['reduced'])

plt.title("Bar chart")

plt.xlabel("Reduced Activity") 

plt.ylabel('Gender')

plt.show()
```

# Add age_category column

```python
df['age_category'] = np.select([

(df['age'] == 0.19),

(df['age'] >=0.20) & (df['age'] <= 0.39), (df['age'] >=0.40) & (df['age'] <= 0.59),

(df['age'] >= 0.60)

], ['Teen', 'Adult', 'Middle Adult', 'Senoir Adult'])
```
